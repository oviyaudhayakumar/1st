```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,15,45])
lables1=["c",'C++','java','python']
myexplode = (0,0,0,0.1)
plt.pie(y,labels = lables1,explode = myexplode, shadow = True )
plt.title("Languages")
plt.show()
```
